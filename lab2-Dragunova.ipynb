```python
import numpy as np
import pandas as pd

# load data with pandas
df = pd.read_csv('cars.csv')
# print(df)

# convert a pandas dataframe to a numpy array
X = df.values[:, 1:-1]
print(type(X))
print(X.shape)
print(X)
```

```python
#Get first row and the second to last row

x1 = X[0, :]
x2 = X[-2, :]

print(x1)
print(x2)

# Addition

y = x1 + x2
print(y)
```

```python
# Get first 5 rows and last 5 rows 

m1 = X[0:5, :]
m2 = X[-5:, :]

print(m1)
print(m2)

# Subtraction
m = m1 - m2
print(m)
```

```python
# this is your code to get two vectors
# the first column 
c1 = X[0:406, 0:1]
#the second to last column 
c2 = X[0:406, -2:-1]

print(c1)
print()
print(c2)
```

```python
import math
# 1. addition and subtraction
print(c1+c2)
print()
print(c1-c2)

# 2. element-wise multiplication
c4 = c1*c2
print(c4)
print()

# 3. inner product
# reshape c2 
c5 = np.empty_like(c2)
c5[:] = c2
c5 = c5.reshape(1,406)
print(np.dot(c1,c5))
print()

# 4. three norms
# norm 1
print("L1 of array 1: ", np.linalg.norm(c1, 1))
print("L2 of array 2: ", np.linalg.norm(c2, 1))
print()

# norm 2
sum1 = 0
for i in c1:
    sum1 += i**2
    
print("L2 of array 1: " , math.sqrt(sum1))

sum2 = 0
for i in c2:
    sum2 += i**2
    
print("L2 of array 2: " , math.sqrt(sum2))
print()

# norm 3
print("L3 of array 1: ", np.linalg.norm(c1, np.inf))
print("L3 of array 2: ", np.linalg.norm(c2, np.inf))
print()


# 5. compute the distance between these two vectors
print("distance: {}".format(np.linalg.norm(c1-c2)))
```

```python
# this is your code to get two matrices: A, B
A= X[0:406, 0:3]
B= X[0:406, -3:]
print(A)
print(B)
```

```python
# 1. addition and subtraction 
print(A+B)
print(A-B)
print()

# 2. compute the transpose of the second matrix: C = B^T, and print the shape of C
print("Shape of B: ", B.shape)
C = np.transpose(B)
print("Shape of C: ", C.shape)
# 3. compute the product between A and C, and print the result and its shape
print(np.dot(A,C).shape)
print()
# 4. compute the transpose of the first matrix: D = A^T, and print the shape of D
print("Shape of A: ", A.shape)
D = np.transpose(A)
print("Shape of D: ", D.shape)
# 5. compute the product between D and B, and print the result and its shape
print(np.dot(B,D).shape)
print()
```
